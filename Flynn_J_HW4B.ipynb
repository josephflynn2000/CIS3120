## Question#1:
### Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt
X = range(1, 50)
Y = [value * 3 for value in X]
print("Values of X:")
print(*range(1,50)) 
print("Values of Y (thrice of X):")
print(Y)

#sets x and y coordinates, then connects the points with a line
plt.plot(X,Y, linestyle='solid')

#adds x-axis label
plt.xlabel("x - axis")

#adds y-axis label
plt.ylabel("y - axis")

#adds title 
plt.title("Draw a line.")

#shows graph
plt.show()
```

## Question#2:
### Write a Python programming to display a bar chart of the popularity of programming Languages.

Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt

#x-labels
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']

#y-values associated to x-values
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

#position of column 
x_pos = [i for i, _ in enumerate(x)]

#plots bars with color fill of blue and edgecolor of black
plt.bar(x_pos, popularity, color='blue', edgecolor ="black")

#labels and titles
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago")

#adds tick marks
plt.xticks(x_pos, x)

# Turn on the grid
#Allows for major grid to have different styles
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')

# Customize the minor grid
plt.grid(which='minor', linestyle='dashdot', linewidth='0.25', color='purple')

#shows graph 
plt.show()
```

## Question#3:
### Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.

```python
#pretty much the same as question 2

import matplotlib.pyplot as plt
x = ['Java', 'Python', 'PHP', 'JS', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]

#barh has the bars on the y-axis. While bar has bars on the x-axis.
plt.barh(x_pos, popularity, color='red', edgecolor ="black")

plt.xlabel("Popularity")
plt.ylabel("Languages")
plt.title("Popularity of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago")
plt.yticks(x_pos, x)
# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Customize the minor grid
plt.grid(which='minor', linestyle='-', linewidth='0.25', color='blue')

#shows graph
plt.show()
```

## Question#5:
### Write a Python programming to create a pie chart of the popularity of programming Languages.

Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt

# Data to plot
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "brown"]

extra = [popuratity[0]/sum(popuratity),popuratity[1]/sum(popuratity),popuratity[2]/sum(popuratity),popuratity[3]/sum(popuratity),popuratity[4]/sum(popuratity),popuratity[5]/sum(popuratity)]

# explodes slices
explode = (extra[0], extra[1],extra[2],extra[3],extra[4],extra[5])  

# Plot
plt.pie(popuratity, explode=explode, labels=languages, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')

#shows graph
plt.show()
```

## Question#6:
###  Population pyramid can be used to show either the distribution of the groups ordered by the volumne. Or it can also be used to show the stage-by-stage filtering of the population as it is used below to show how many people pass through each stage of a marketing funnel.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data
df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/email_campaign_funnel.csv")

# Draw Plot
plt.figure(figsize=(13,10), dpi= 80)
group_col = 'Gender'
order_of_bars = df.Stage.unique()[::-1]
colors = [plt.cm.Spectral(i/float(len(df[group_col].unique())-1)) for i in range(len(df[group_col].unique()))]

for c, group in zip(colors, df[group_col].unique()):
    sns.barplot(x='Users', y='Stage', data=df.loc[df[group_col]==group, :], order=order_of_bars, color=c, label=group)

  
 #adds x-axis title and has selected font and italized
plt.xlabel("Users", style='italic', font = 'Times New Roman')

 #adds y-label and increases the font size
plt.ylabel("Stage of Purchase", fontsize = 12)

plt.yticks(fontsize=12)

 #adds title and gives it a color
plt.title("Population Pyramid of the Marketing Funnel", fontsize=22, color='blue')

 #adds a legend and sets location
plt.legend(loc = 'right')

 #shows graph
plt.show()
    
```
