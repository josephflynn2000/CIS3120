## Question#1:
### Complete the Python program below to display the name of the most recently added dataset and collect the total number of datasets currently listed on the URL provided below.
Note: You may need to install **cssselect** module

```python
!pip install cssselect
```

```python
from lxml import html
import requests
response = requests.get('http://catalog.data.gov/dataset?q=&sort=metadata_created+desc')

#html.fromstring converts the http response into an html element
doc = html.fromstring(response.text)
#a specifc element is then selected with cssselect(). In this case, its the content under the h3 head, under class of dataset-heading.
#The .text_content coverts the element to text
title = doc.cssselect('h3.dataset-heading')[0].text_content()


print("The name of the most recently added dataset on data.gov:")
#strips the title and prints it
print(title.strip())

response1 = requests.get('http://www.data.gov/')

#html.fromstring converts the http response into an html element
doc_gov = html.fromstring(response1.text)
#a specifc element is then selected with cssselect(). Then it selects the element at samll a and the first index.
link_gov = doc_gov.cssselect('small a')[0]

print("Number of datasets currently listed on data.gov:")
#changes specific element to text before printing
print(link_gov.text)
```

**OUTPUT TEXT: **
The name of the most recently added dataset on data.gov:
Canterbury, New Zealand 2018
Number of datasets currently listed on data.gov:
224,654 datasets

## Question#3:
### Complete the Python program below to extract and display all the header tags from the URL below.
Note: HTML headers are 'h1', 'h2','h3','h4','h5','h6'


```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

#saves data from url
html = urlopen('https://blogs.baruch.cuny.edu/abuk/')
#turns data to html.parser
bs =  BeautifulSoup(html, 'html.parser') 
#setup empty list to save headers to later
titles = []

#loop that finds headers 1 to 6
for found in bs.find_all(['h1', 'h2','h3','h4','h5','h6']):
    #takes found header and stores it in list named titles
    titles.append(found)

#prints title list to have each header spaced by one line
print('List all the header tags :', *titles, sep='\n\n')
```
